8. What is the binomial distribution and how is it used in probabilty?

9. Write a python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = [(x - mean) / std_dev for x in data]

    # Plotting the Z-scores
    plt.figure(figsize=(8, 6))
    plt.hist(z_scores, bins=30, edgecolor='black', alpha=0.7)
    plt.title('Z-scores Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage:
data = [10, 12, 10, 15, 14, 13, 12, 15, 16, 11, 14, 17, 10, 13]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.


```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem(non_normal_distribution, sample_size, num_samples):
    sample_means = []

    # Take 'num_samples' samples, each of size 'sample_size'
    for _ in range(num_samples):
        sample = np.random.choice(non_normal_distribution, size=sample_size, replace=True)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Plot the distribution of sample means
    plt.figure(figsize=(8, 6))
    plt.hist(sample_means, bins=30, edgecolor='black', alpha=0.7, density=True)
    plt.title(f"Distribution of Sample Means\nSample Size = {sample_size}, Number of Samples = {num_samples}")
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return sample_means

# Example usage:
# Create a non-normal distribution (Uniform distribution between 0 and 10)
non_normal_dist = np.random.uniform(0, 10, 10000)

# Set sample size and number of samples
sample_size = 50   # Size of each sample
num_samples = 1000 # Number of samples to take

# Apply the Central Limit Theorem
sample_means = central_limit_theorem(non_normal_dist, sample_size, num_samples)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a given data point.
    """
    z_score = (x - mean) / std_dev
    return z_score

def compare_to_standard_normal_distribution(data_point, mean, std_dev):
    # Calculate the Z-score for the data point
    z_score = calculate_z_score(data_point, mean, std_dev)
    print(f"Z-score for the data point {data_point} is: {z_score:.4f}")

    # Generate x values for plotting the standard normal distribution (mean=0, std=1)
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals, loc=0, scale=1)  # Standard Normal Distribution

    # Plot the standard normal distribution
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, label="Standard Normal Distribution", color='blue')

    # Plot a vertical line for the calculated Z-score
    plt.axvline(z_score, color='red', linestyle='dashed', label=f"Z-score = {z_score:.2f}")

    plt.title(f"Comparison of Z-score with Standard Normal Distribution\nData Point: {data_point}")
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
data_point = 85    # Example data point
mean = 70          # Mean of the distribution
std_dev = 10       # Standard deviation of the distribution

compare_to_standard_normal_distribution(data_point, mean, std_dev)
```

15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, significance_level=0.05):
    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample_data)
    n = len(sample_data)

    # Step 2: Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Step 3: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Step 4: Make a decision
    if p_value < significance_level:
        decision = "Reject the null hypothesis"
    else:
        decision = "Fail to reject the null hypothesis"

    # Output results
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    return z_statistic, p_value, decision

# Example usage:
sample_data = [22, 23, 21, 25, 27, 24, 22, 26, 23, 22]  # Sample data
population_mean = 23  # Known population mean
population_std = 2    # Known population standard deviation
alpha = 0.05          # Significance level

z_statistic, p_value, decision = z_test(sample_data, population_mean, population_std, alpha)

```

16. Create  a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(sample_data, confidence_level=0.95):
    # Step 1: Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (Bessel's correction)
    n = len(sample_data)

    # Step 2: Calculate the standard error of the mean
    se = sample_std / np.sqrt(n)

    # Step 3: Find the critical value (Z for normal distribution)
    z_score = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # Z value for 95% CI is approximately 1.96

    # Step 4: Calculate the margin of error
    margin_of_error = z_score * se

    # Step 5: Calculate the confidence interval
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Standard Deviation: {sample_std:.4f}")
    print(f"Standard Error: {se:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")

    # Plotting the confidence interval
    plt.figure(figsize=(8, 6))
    plt.plot([0, 1], [sample_mean, sample_mean], color='blue', label=f"Sample Mean = {sample_mean:.4f}")
    plt.plot([0, 1], [ci_lower, ci_lower], color='red', linestyle='dashed', label=f"Lower CI = {ci_lower:.4f}")
    plt.plot([0, 1], [ci_upper, ci_upper], color='red', linestyle='dashed', label=f"Upper CI = {ci_upper:.4f}")
    plt.fill_between([0, 1], ci_lower, ci_upper, color='skyblue', alpha=0.3)
    plt.title(f"Confidence Interval (95%) for the Sample Mean")
    plt.legend()
    plt.grid(True)
    plt.show()

    return ci_lower, ci_upper

# Example usage
sample_data = [22, 23, 21, 25, 27, 24, 22, 26, 23, 22]  # Sample dataset
confidence_interval(sample_data)
```

23. Implememt Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean=50, Std=15

# Step 2: Compute Mean and Standard Deviation
mean = np.mean(data)
std_dev = np.std(data)

# Step 3: Apply Z-score normalization
z_scores = (data - mean) / std_dev

# Step 4: Plot before and after normalization
plt.figure(figsize=(12, 5))

# Original Data Histogram
plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color="blue", edgecolor="black")
plt.axvline(mean, color="red", linestyle="dashed", label=f"Mean={mean:.2f}")
plt.title("Original Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

# Normalized Data Histogram
plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, kde=True, color="green", edgecolor="black")
plt.axvline(0, color="red", linestyle="dashed", label="Mean=0 (Normalized)")
plt.title("Z-score Normalized Data Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()

# Step 5: Print Mean & Standard Deviation after Normalization
print(f"Original Mean: {mean:.2f}, Original Std Dev: {std_dev:.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Normalized Std Dev: {np.std(z_scores):.2f}")
```
